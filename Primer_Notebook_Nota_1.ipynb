# Problema de Ingeniería (Nivel Básico)

**Calcular el volumen de agua almacenado en un tanque rectangular.**

Este tipo de cálculo es común en:

* Reservorios de riego
* Tanques de almacenamiento
* Sistemas hidráulicos

La fórmula física es:

[
Volumen = Largo \times Ancho \times Altura
]

---

# 1. Definición del algoritmo

Antes de programar, siempre debemos definir:

### Entradas (Inputs)

Datos que el usuario debe ingresar.

* largo del tanque (m)
* ancho del tanque (m)
* altura del agua (m)

### Proceso (Process)

Operación matemática que realiza el programa.

[
V = largo \times ancho \times altura
]

### Salida (Output)

* volumen de agua almacenado en **m³**

---

# 2. Diseño del algoritmo paso a paso

1. Iniciar programa
2. Pedir al usuario el **largo del tanque**
3. Pedir el **ancho del tanque**
4. Pedir la **altura del agua**
5. Calcular el volumen
6. Mostrar el volumen calculado
7. Finalizar

---

# 3. Pseudocódigo

Antes de escribir Python, se recomienda escribir pseudocódigo.

```
INICIO

leer largo
leer ancho
leer altura

volumen = largo * ancho * altura

mostrar volumen

FIN
```

---

# 4. Código en Python

```python
# --------------------------------------------
# Programa: Cálculo del volumen de un tanque
# Aplicación: Ingeniería hidráulica básica
# --------------------------------------------

# Entrada de datos
largo = float(input("Ingrese el largo del tanque (m): "))
ancho = float(input("Ingrese el ancho del tanque (m): "))
altura = float(input("Ingrese la altura del agua (m): "))

# Proceso
volumen = largo * ancho * altura

# Salida
print("El volumen de agua almacenado es:", volumen, "m3")
```

---

# 5. Ejemplo de ejecución

Si el usuario ingresa:

```
largo = 10 m
ancho = 5 m
altura = 2 m
```

El programa calcula:

[
V = 10 \times 5 \times 2 = 100
]

Resultado:

```
El volumen de agua almacenado es: 100 m3
```
