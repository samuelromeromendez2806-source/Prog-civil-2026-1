```markdown
## SESIÃN 1: IntroducciÃģn a los Procesos Iterativos

**ÂŋPor quÃĐ necesitamos ciclos?**
Imagina que debes calcular el promedio de temperatura durante los 7 dÃ­as de la semana. Sin usar ciclos, tendrÃ­as que escribir lÃ­neas de cÃģdigo para pedir la temperatura siete veces distintas. Un ciclo (o bucle) nos permite escribir esa instrucciÃģn una sola vez y decirle a Python que la repita automÃĄticamente.

### 1. El Bucle `while` (Mientras que...)
La estructura `while` ejecuta un bloque de cÃģdigo continuamente *mientras* una condiciÃģn siga siendo verdadera.

Para evitar que el ciclo se repita infinitamente, utilizamos un **contador**: una variable que aumenta su valor en cada iteraciÃģn para llevar el control.

```

### Celda 3: CÃģdigo (Python)

````
# Ejemplo de bucle while controlado por un contador
contador_dias = 1

# El ciclo se ejecutarÃĄ mientras el contador sea menor o igual a 5
while contador_dias <= 5:
    print("Registrando datos para el dÃ­a nÃšmero:", contador_dias)
    # Aumentamos el contador en 1 para que el ciclo avance
    contador_dias = contador_dias + 1
````

```markdown
## SESIÃN 2: AplicaciÃģn PrÃĄctica con Acumuladores

En ingenierÃ­a no solo contamos eventos, tambiÃĐn sumamos cantidades a lo largo del tiempo. Para esto usamos un **acumulador**: una variable que suma o "acumula" valores progresivamente durante el ciclo.

### Mini-proyecto 1: Promedio de Mediciones TÃĐrmicas
Vamos a automatizar la recolecciÃģn de 5 mediciones de temperatura en un proceso agroindustrial y calcularemos su promedio.

```

### Celda 8: CÃģdigo (Python)

````
# Programa para promediar 5 mediciones de temperatura
suma_temperaturas = 0  # Inicializamos el acumulador en cero

for i in range(5):
    # Se repite 5 veces la peticiÃģn de datos al usuario
    temp_ingresada = float(input("Ingrese la temperatura medida (Â°C): "))

    # Acumulamos el valor ingresado a la suma total
    suma_temperaturas = suma_temperaturas + temp_ingresada

# Al terminar el ciclo, calculamos y mostramos el promedio
promedio = suma_temperaturas / 5
print("\nLa temperatura promedio del proceso es:", promedio, "Â°C")
````

```markdown
### Mini-proyecto 2: Registro HidrolÃģgico
Este programa simula el registro de la precipitaciÃģn diaria durante una semana en una estaciÃģn meteorolÃģgica local, calculando el total de lluvia acumulada, un dato vital en el diseÃąo de drenajes agrÃ­colas o civiles.
```

### Celda 10: CÃģdigo (Python)

````
# Programa de registro de precipitaciÃģn semanal
lluvia_total = 0  # Nuestro acumulador

for dia in range(7):
    # Usamos (dia + 1) para que en pantalla se vea "dÃ­a 1" en lugar de "dÃ­a 0"
    lluvia_diaria = float(input(f"Ingrese la lluvia registrada el dÃ­a {dia + 1} (en mm): "))
    lluvia_total = lluvia_total + lluvia_diaria

print("\nLa precipitaciÃģn total acumulada en la semana es de:", lluvia_total, "mm")
````
