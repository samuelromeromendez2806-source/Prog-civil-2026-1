# **Teoría del Confort Térmico en Ganado Vacuno**

El confort térmico en el ganado vacuno se refiere al estado en el que los animales no experimentan estrés por calor ni por frío. Mantener a los animales en su zona de confort térmico es crucial para su bienestar, salud y productividad.

## **Factores que Afectan el Confort Térmico:**

1.  **Temperatura Ambiental:** La temperatura del aire es un factor directo que afecta la pérdida o ganancia de calor del animal.
2.  **Humedad Relativa:** La alta humedad dificulta la disipación de calor por evaporación (sudoración y jadeo), haciendo que las temperaturas elevadas sean aún más estresantes.
3.  **Radiación Solar:** La exposición directa al sol aumenta significativamente la carga de calor en los animales.
4.  **Velocidad del Viento:** El viento puede ayudar a disipar el calor del cuerpo del animal, especialmente en ambientes cálidos, pero puede aumentar la pérdida de calor en ambientes fríos.
5.  **Precipitación:** La lluvia puede mojar el pelaje del animal, alterando su capacidad de aislamiento térmico.

## **Impacto del Estrés Térmico:**

Cuando el ganado vacuno se encuentra fuera de su zona de confort térmico, experimenta estrés térmico, que puede manifestarse de las siguientes maneras:

*   **Disminución de la Producción de Leche:** Las vacas lecheras reducen su consumo de alimento y, por ende, su producción de leche.
*   **Menor Ganancia de Peso:** El ganado de engorde consume menos alimento y utiliza energía para combatir el estrés, resultando en una menor ganancia de peso.
*   **Problemas Reproductivos:** El estrés térmico puede afectar negativamente la fertilidad, provocando una disminución en las tasas de concepción y un aumento en las pérdidas embrionarias.
*   **Alteraciones en el Comportamiento:** Los animales buscan sombra, aumentan su frecuencia respiratoria (jadeo) y reducen su actividad.
*   **Mayor Susceptibilidad a Enfermedades:** El sistema inmunológico de los animales se debilita, haciéndolos más vulnerables a enfermedades.
*   **Mortalidad:** En casos severos y prolongados, el estrés térmico puede llevar a la muerte de los animales.

## **Índices de Confort Térmico:**

Para cuantificar el grado de estrés térmico, se utilizan diversos índices que combinan múltiples factores ambientales. El **Índice de Temperatura y Humedad (ITH)** es uno de los más comunes y se calcula a partir de la temperatura de bulbo seco y la temperatura de bulbo húmedo. Este índice ayuda a los productores a identificar situaciones de riesgo y tomar medidas preventivas para mitigar los efectos negativos del estrés térmico.

# **Métodos y cálculos**
• Índice de temperatura y humedad (producción intensiva)


$𝐼𝑇𝐻 = 0.72(𝑇𝑏𝑠 + 𝑇𝑏ℎ) + 40.6 [1] $

$ T _ {bs}$: Temperatura bulbo seco (°C)

$ T _ {bh}$: Temperatura de bulbo húmedo (°C)

Para la correlación de los datos de la especie bovina en la producción de leche se recomienda
utilizar los límites propuestos por National Weather Service:


Sí $ITH ≤ 70 $ ambiente no estresante
Sí $ITH > 70 \space and \space ≤ 78$ situación de alerta por estrés térmico
Sí $ITH > 78 \space and \space ≤ 83 $ peligro por estrés térmico
Sí $ITH > 83$ situación crítica


````python
 #ENTRADA: pedir tbs (termperatura del bulbo seco) y tbh (Tempertatura del bulbo humedo)
 tbs = input("Ingrese la temperatura del bulbo seco: ")
 tbh = input("Ingrese la temperatura del bulbo húmedo: ")

 #PROCESO: Calculo de ith (indice de temperatura y humedad)
 ith= 0.72 * (float(tbs) + float(tbh)) + 40.6

 #SALIDA: imprimir resultado de la evaluacion de ith y el confort termico
 print("El indice de temperatura y humedad es: ", ith)
 print("El confort térmico es: ")

 if ith <= 70:
  print("Ambiente no estresante")
 elif ith > 70 and ith <= 78:
  print("Situación de alerta por estrés térmico")
 elif ith > 78 and ith <= 83:
  print("Peligro por estrés térmico")
 else:
  print("Situación crítica")
````
