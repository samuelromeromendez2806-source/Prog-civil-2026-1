**Dosificacion de mezcla para columnas**

# **Planteamiento**:

Un ingeniero residente necesita calcular cuántos sacos de cemento se requieren para vaciar una columna de sección circular. Se sabe que para un concreto estándar de 210 kg/cm2(3000 psi), se consumen 9.5 sacos de cemento por cada metro cúbico de mezcla.

**Requerimientos**:

1. Solicitar la altura de la columna en metros.
2. Solicitar el diámetro de la columna en metros.
3. Calcular el volumen de la columna: $$V= \pi * r^2 * h$$
. (Sugerencia: usar 3.1416
 o importar math.pi).
4. Calcular la cantidad de sacos: Sacos=
$$sacos= v * 9.5$$
.
Prueba del programa:

Para una columna de 3 metros de alto y 0.4 metros de diámetro (r=0.2
):

Volumen ≈0.377 m3
.
Sacos necesarios ≈3.58

 (el estudiante verá aquí la importancia de los tipos decimales).


````python
# ENTRADA Solicitar la altura de la columna en metros.
# Solicitar el diámetro de la columna en metros.
h = float(input("Ingrese la altura de la columna en metros: "))
d = float(input("Ingrese el diámetro de la columna en metros: "))

# PROCESO: Calcula el radio r, calcula el volumen de la columna, calcula sacos
r = d/2
v = 3.1416 * r**2 * h
sacos = v * 9.5

# SALIDA: Imprime el resultado
print(f"Volumen ≈{v:.3f} m3")
print(f"Sacos necesarios ≈{sacos:.2f}")

````

```markdown
Aquí está la explicación de cada línea del código:

```python
h = float(input("Ingrese la altura de la columna en metros: "))

```
Estas dos líneas solicitan al usuario la `altura` (h) y el `diámetro` (d) de la columna. `input()` lee la entrada como una cadena de texto, y `float()` la convierte a un número decimal para poder realizar cálculos con ella.

```python

```
Esta línea calcula el `radio` (r) de la columna dividiendo el `diámetro` (d) ingresado por el usuario entre 2.

```python
v = 3.1416 * r**2 * h
```
Esta línea calcula el `volumen` (v) de la columna cilíndrica utilizando la fórmula `V = π * r^2 * h`. Aquí, `3.1416` se usa como una aproximación de Pi, `r**2` calcula el radio al cuadrado, y `h` es la altura.

```python

```
Esta línea calcula la cantidad total de `sacos` de cemento necesarios. Multiplica el `volumen` (v) de la columna por `9.5`, que es la cantidad de sacos de cemento requeridos por cada metro cúbico de mezcla.
